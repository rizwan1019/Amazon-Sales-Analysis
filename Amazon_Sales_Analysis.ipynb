```python
# importing python libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading csv file from device

df = pd.read_csv('/content/amazon.csv')

# top 5 rows
df.head()
```

```python
# Transposing datasets from rows to columns

df.head().T
```

```python
# Shape of dataset

df.shape
```

```python
# Checking datatypes and null columns

df.info()
```

```python
# checking null values

df.isna().sum()
```

```python
# filling null value in particular columns

df['rating_count'].fillna(0,inplace=True)
```

```python
# Checking null values again after filling

df.isna().sum()
```

```python
# replacing ₹ and ,

df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True)
df['actual_price']
```

```python
#  replacing ₹ and ,

df['discounted_price'] = df['discounted_price'].replace('[₹,]', '', regex=True)
df['discounted_price']
```

```python
# replacing %

df['discount_percentage']=df['discount_percentage'].replace('%', '', regex=True)
df['discount_percentage']
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','',regex=True)
df['rating_count']
```

```python
df['discounted_price']=df['discounted_price'].astype(float)
```

```python
df['actual_price']=df['actual_price'].astype(float)
```

```python
df['discount_percentage']=df['discount_percentage'].astype(float)
```

```python
df['rating_count']=df['rating_count'].astype(float)
```

```python
df['rating']=pd.to_numeric(df['rating'],errors='coerce')
df['rating']
```

```python
df.info()
```

```python
df.describe()
```

```python
# 1. How are product prices distributed, and are there notable outliers?

sns.scatterplot(df['actual_price'],color='purple')
plt.title('Distributed product price')
plt.xlabel('Total rows')
plt.ylabel('Actual Price')
plt.show()
```

```python
# Checking max value

df['actual_price'].max()
```

```python
# 2. What is the average rating across different product categories?

df["df_category"] = df['category'].str.split('|',expand=True)[0]
```

```python
df_avg_rating1 = df.groupby('df_category')['rating'].mean().reset_index()
```

```python
sns.lineplot(df_avg_rating1,x='df_category',y='rating',color='red',marker='*',markersize=7)
plt.xticks(rotation=75)
plt.title('Average rating across different product categories')
plt.show()
```

```python
# 3. Is there a correlation between discount percentage and product rating?
plt.figure(figsize=(10,6))
sns.lineplot(df,x='rating',y='discount_percentage',color='green')
plt.title('Correlation between discount percentage and product rating')
plt.show()
```

```python
# 4. What are the top 5 most reviewed products, and what insights do their reviews offer?

top_5_most_review = df.sort_values(by='rating_count',ascending=False).head()
```

```python
top_5_most_review[['product_name','rating','rating_count',]]
```

```python
# 5. Are certain product categories more frequently discounted?


discount_frequency = df.groupby('category')['discount_percentage'].count().reset_index()

discount_frequency.columns = ['category', 'discount_frequency']
```

```python
sorted_by_dis_freq = discount_frequency.sort_values(by='discount_frequency',ascending=False).head(15)
sns.barplot(sorted_by_dis_freq,x='category',y='discount_frequency')
plt.title('product categories more frequently discounted')
plt.xticks(rotation=90)
plt.show()
```

```python
# 6. Do higher-priced products tend to receive higher ratings?

higher_price = df.sort_values(by='actual_price',ascending=False).head(10)
```

```python
higher_price.plot(x='actual_price',y='rating',color='orange')
```

```python
# 7. Is there a relationship between description length and price or rating?
df['description_length']=df['about_product'].str.len()
df['description_length']
```

```python
sns.scatterplot(x='description_length',y='actual_price',data=df)
plt.title('Relationship between description length and price')
plt.show()
```

```python
sns.scatterplot(x='description_length',y='rating',data=df,color='red')
plt.title('Relationship between description length and rating')
plt.show()
```

```python
# 8. How do ratings differ between products with high and low review counts?

high_rated_product = df.sort_values(by='rating_count',ascending=False).head(10)
high_rated_product
```

```python
# high_rated_product.plot(y='rating_count',x='rating',kind='line')
sns.lineplot(high_rated_product,x='rating',y='rating_count',marker='*',markersize=10)
```

```python
low_rated_product = df.sort_values(by='rating_count',ascending=True).head(10)
low_rated_product
```

```python
sns.lineplot(low_rated_product,x='rating',y='rating_count',marker='*',markersize=10)
```

```python
# 10. Are there common patterns in products receiving negative reviews?

low_rating = df[df['rating'] <3]
low_rating
```

```python
sns.lineplot(low_rating,x='rating',y='rating_count',marker='*',markersize=10)
plt.title('Common patterns in products receiving negative reviews')
plt.show()
```

```python
# 12. How does the discount percentage vary across different brands?

# Extracting the brand name
df['brand'] = df['product_name'].str.split(' ',expand=True)[0]
df['brand']
```

```python
# Average percentage of the discount percentage

avg_percentage = df.groupby('brand')['discount_percentage'].mean().head(10)
avg_percentage
```

```python
avg_percentage.plot(kind='bar',x='product_name',y='discount_percentage',color='purple',alpha=1,edgecolor='black')
plt.title('Discount percentage vary across different brands')
plt.xticks(rotation=75)
plt.show()
```

```python
# 16. Can product rating be predicted based on features and category?

df['new_category'] = df['category'].str.split('|',expand=True)[0]
df['new_category']
```

```python
avg_rating = df.groupby('new_category')['rating'].mean().reset_index()
avg_rating
```

```python
# avg_rating.plot(kind='bar',x='new_category',y='rating')
sns.barplot(avg_rating,x='new_category',y='rating', palette='viridis',hue='rating')
plt.title('Product rating predicted based on category')
plt.xticks(rotation=75)
plt.show()
```

```python
# 17. How can insights from this analysis improve product recommendations?

df_sort_rating = df.sort_values(by='rating',ascending=False).head(10)
df_sort_rating
```

```python
df_sort_rating.plot(kind='line',x='brand',y='rating',marker='*',markersize=7,color='purple')
plt.title('Product analysis on the basis of rating to improve product recommendations')
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.show()
```

19. What additional data sources could enhance this analysis?

    To make your analysis even better, you could add some extra information:

      1. Sales Numbers: Knowing how many units of each product were sold would help you see if discounts and good reviews lead to more sales.

      2. Customer Details: Information about who is buying (like their age or where they live) could show you if certain products are popular with specific groups of people.

      3. Competitor Info: Data on similar products from other brands could help you see how your products stack up against the competition.

      4. Marketing Spend: Knowing how much was spent on ads for each product could help you see if marketing efforts are paying off.

      5. Product Lifespan: Knowing how long these products last could help you understand if they are durable or need frequent replacement.

      6. Inventory and Stock: Information on how often products go out of stock could help you see if supply issues are affecting sales.
